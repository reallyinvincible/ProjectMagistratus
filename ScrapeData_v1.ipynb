### Collecting Tweets since 3rd Jan,2020

```python
#search_words = "#murder"
date_since = "2020-03-01"

# Collect tweets
tweets = tw.Cursor(api.search,q="#* -filter:retweets",lang="en",since=date_since).items(10)


tweet_details = [[tweet.user.screen_name, tweet.user.location, tweet.text] for tweet in tweets]
tweet_data = pd.DataFrame(data=tweet_details, columns=['User', "Location","Tweet"])

tweet_data
```

### Tweepy allows an extraction of 100 tweets in 70 second, extracting a total of 10000 tweets, with a appropriate sleep time

```python
import time

#search_words = "#murder"
date_since = "2020-03-03"

# Collect tweets
'''If I want a set of 50000 tweets, '''
count=0
while count!=10000:
    print("Count: ",count,"  : ")
    count=count+1;
    tweets = tw.Cursor(api.search,q="#* -filter:retweets",lang="en",since=date_since).items(10)
    tweet_details0 = [[tweet.user.screen_name, tweet.user.location, tweet.text] for tweet in tweets]
    tweet_details.extend(tweet_details0)
    print(len(tweet_details),"\n")
    time.sleep(7) #Sleep loop for 70 sec estimating 100 tweeys in 70 seconds



```

```python
tweet_data = pd.DataFrame(data=tweet_details, columns=['User', "Location","Tweet"])
tweet_data
```

```python
tweet_data.to_csv('tweet_data1.csv') 
```

```python
import json
with open('data.json', 'w') as outfile:
    json.dump({'contentItems': tweet_details}, outfile,indent=4)
```

### Converting all tweets thus collected into a IBM Watson acceptable JSON format

```python
#create a list of dictionaries to later import it into json format that is accepted by Watson
df=tweet_data
d={}#The dictionary
dlist=[]#The list to append the dictionary to

for i in df.itertuples():
    d={}
    d['content']=i[3]
    d['contenttype']="text/plain"
    d['created']= 0
    d['id']= i[0]
    d['language']= "en"
    dlist.append(d)
    print(d,"\n")
    
```

```python
with open('data.json', 'w') as outfile:
    json.dump({'contentItems': dlist}, outfile,indent=4)
```
